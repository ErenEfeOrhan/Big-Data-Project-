# Adım 1 - Docker Ortamının Kurulumu

Bu notebook proje sunumu ve teslimi için ilgili adımı özetler.


## Amaç

Bu adımda tüm servislerin Docker ve Docker Compose ile izole, tekrar çalıştırılabilir bir ortamda ayağa kaldırılması hedeflenmiştir.


## Kapsanan Servisler

- Zookeeper
- Kafka Broker
- Spark / Jupyter
- MLflow
- Python Kafka Producer


## İlgili Dosyalar

- docker-compose.yml
- zookeeper/Dockerfile
- kafka/Dockerfile
- spark/Dockerfile
- producer/Dockerfile
- mlflow/Dockerfile


## Çalıştırma Komutları

PowerShell üzerinde:

```powershell
docker compose build
docker compose up -d zookeeper kafka spark mlflow
docker compose --profile producer up -d producer
docker ps
```


## Sunumda Gösterilecek Kanıt

- docker ps çıktısı
- Çalışan container listesi
- GitHub üzerinde Dockerfile ve docker-compose.yml dosyaları
