# Adım 2 - Kafka ile Streaming Veri Üretimi

Bu notebook proje sunumu ve teslimi için ilgili adımı özetler.


## Amaç

Spotify CSV verisini gerçek zamanlı veri akışı gibi Kafka topic'ine JSON formatında göndermek.


## Producer Mesaj İçeriği

Producer her mesajda şu alanları üretir:

- event_time
- event_type
- user_id
- track_id
- artists
- track_name
- popularity
- audio feature kolonları


## Hız Ayarı

Mesaj gönderim hızı MESSAGES_PER_SECOND environment değişkeni ile kontrol edilir.


## Kontrol Komutları

```powershell
docker compose --profile producer up -d producer
docker logs spotify_producer --tail 30
```


## Sunumda Anlatılacak Nokta

Kafka burada gerçek zamanlı veri akışını simüle eder. CSV satırları producer tarafından olay/event formatına çevrilerek Kafka'ya gönderilir.
