# Adım 3 - Spark Structured Streaming ve Delta Lake

Bu notebook proje sunumu ve teslimi için ilgili adımı özetler.


## Amaç

Kafka'dan gelen JSON mesajlarını Spark Structured Streaming ile okuyup Delta Lake formatında Bronze, Silver ve Gold katmanlarına yazmak.


## Katmanlar

- Bronze: Kafka'dan gelen ham veri
- Silver: Parse edilmiş, temizlenmiş ve tip dönüşümleri yapılmış veri
- Gold Analytics: Analitik özetler
- Gold Features: ML modelleri için feature tablosu


## İlgili Spark Job Dosyaları

- spark/jobs/stream_kafka_to_bronze.py
- spark/jobs/stream_bronze_to_silver.py
- spark/jobs/stream_silver_to_gold.py


## Çalıştırma Örneği

```powershell
docker exec -e RUN_SECONDS=180 spotify_spark spark-submit --packages "org.apache.spark:spark-sql-kafka-0-10_2.12:3.5.0,io.delta:delta-spark_2.12:3.2.0" --conf "spark.sql.extensions=io.delta.sql.DeltaSparkSessionExtension" --conf "spark.sql.catalog.spark_catalog=org.apache.spark.sql.delta.catalog.DeltaCatalog" /home/jovyan/work/spark/jobs/stream_kafka_to_bronze.py
```


## Delta Kontrol Komutları

```powershell
dir delta\bronze\spotify_tracks\_delta_log
dir delta\silver\spotify_tracks\_delta_log
dir delta\gold\spotify_analytics\_delta_log
dir delta\gold\spotify_tracks_features\_delta_log
```


## Sunumda Anlatılacak Nokta

Bu adım veri mühendisliği pipeline'ının merkezidir. Kafka'dan gelen veri Spark ile işlenir ve Delta Lake üzerinde katmanlı veri mimarisi oluşturulur.
